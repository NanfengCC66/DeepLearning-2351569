```python
import os
import torch
import torch.nn as nn
import torch.utils.data as Data
import torchvision
import torch.nn.functional as F
import numpy as np

# 超参数
learning_rate = 1e-4
keep_prob_rate = 0.7 
max_epoch = 3
BATCH_SIZE = 50

# 加载数据
train_data = torchvision.datasets.MNIST(root='./mnist/', train=True, transform=torchvision.transforms.ToTensor(), download=True)
train_loader = Data.DataLoader(dataset=train_data, batch_size=BATCH_SIZE, shuffle=True)
test_data = torchvision.datasets.MNIST(root='./mnist/', train=False)

test_x = torch.unsqueeze(test_data.data, dim=1).type(torch.FloatTensor)[:2000]/255. 
test_y = test_data.targets[:2000].numpy()

class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        # 第一层卷积
        self.conv1 = nn.Sequential(
            nn.Conv2d(1, 32, 7, 1, 3), # in=1, out=32, kernel=7, padding=3
            nn.ReLU(),
            nn.MaxPool2d(2),
        )
        # 第二层卷积
        self.conv2 = nn.Sequential(
            nn.Conv2d(32, 64, 5, 1, 2), # in=32, out=64, kernel=5, padding=2
            nn.ReLU(),
            nn.MaxPool2d(2),
        )
        self.out1 = nn.Linear(7 * 7 * 64, 1024)
        self.dropout = nn.Dropout(1 - keep_prob_rate) 
        self.out2 = nn.Linear(1024, 10)

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = x.view(-1, 7 * 7 * 64) 
        x = F.relu(self.out1(x))
        x = self.dropout(x)
        return self.out2(x)

def train():
    cnn = CNN()
    optimizer = torch.optim.Adam(cnn.parameters(), lr=learning_rate)
    loss_func = nn.CrossEntropyLoss()
    
    print("PyTorch 训练开始...")
    for epoch in range(max_epoch):
        for step, (x_, y_) in enumerate(train_loader):
            output = cnn(x_)  
            loss = loss_func(output, y_)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            if step % 200 == 0:
                cnn.eval()
                with torch.no_grad():
                    test_out = cnn(test_x)
                    pred_y = torch.max(test_out, 1)[1].numpy()
                    accuracy = np.sum(pred_y == test_y) / test_y.size
                    print(f"Epoch {epoch}, Step {step}, Accuracy: {accuracy:.4f}")
                cnn.train()

if __name__ == '__main__':
    train()
```
